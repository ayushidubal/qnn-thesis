```python
import matplotlib.pyplot as plt
```

```python
def get_scores(dataset, score, fm):
    sf = open(f"{dataset[0]}/{score}.txt")
    flag = False
    for line in sf.readlines():
        if flag:
            scores_str = line[:-3].split(', ')
            break
        if line.startswith(fm):
            flag = True
            continue
    scores = []
    for s in scores_str:
        if s == 'None':
            scores.append(0)
        else:
            scores.append(float(s))
    return scores
```

```python
eqs = ['def', '8', '8.1', '8.2', '8.3', '8.4', '9', '9.1', '9.2', '10', '10.1', '10.2', '11', '11.1', '11.2', '12', '12.1', '12.2', 'lin', 'rbf', 'poly']
```

```python
cacc = get_scores("Raisin", "Acc", "Lin")
crec = get_scores("Raisin", "Rec", "Lin")
cf1 = get_scores("Raisin", "F1", "Lin")
cprec = get_scores("Raisin", "Prec", "Lin")
```

```python
razz = get_scores("Raisin", "Acc", "ZZ")
plt.bar(eqs, razz+cacc)
plt.title("Raisin ZZ Accuracy")
```

```python
razz = get_scores("Raisin", "Prec", "ZZ")
plt.bar(eqs, razz)
plt.title("Raisin ZZ Prec")
```

```python
hazz = get_scores("Heart", "Acc", "ZZ")
plt.bar(eqs, hazz)
plt.title("Heart ZZ Accuracies")
```

```python
raz7 = get_scores("Raisin", "Acc", "ZZZZZZZ")
plt.bar(eqs, raz7)
plt.title("Raisin ZZZZZZZ Accuracies")
```
